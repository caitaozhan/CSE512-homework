```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)
y = np.exp(x)

plt.figure()
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$\exp(x)$')
plt.show()
```

```python
plt.figure()
plt.plot(x, np.exp(-x))
plt.xlabel('$x$')
plt.ylabel('$\exp(-x)$')
plt.show()
```
