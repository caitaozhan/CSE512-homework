```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(-1, 1, 100)
y = np.exp(x)

plt.figure()
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$\exp(x)$')
plt.show()
```

```python
plt.figure()
plt.plot(x, np.exp(-x))
plt.xlabel('$x$')
plt.ylabel('$\exp(-x)$')
plt.show()
```

```python
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
def_figsize = 56, 20
#rcParams['figure.figsize'] = 28, 20

plt.style.use('seaborn-paper')
sns.set(context='paper', font_scale=1.6)
sns.set_style("whitegrid")
sns.set_palette(sns.xkcd_palette(['black', 'greyish']))

SMALL_SIZE = 40
MEDIUM_SIZE = 40
BIGGER_SIZE = 40
#
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE) # fontsize of the figure title

df_1 = pd.read_csv('plot_data/p1.csv', header=None)
df_2 = pd.read_csv('plot_data/p2.csv', header=None)
df_3 = pd.read_csv('plot_data/p3.csv', header=None)

Y_1 = df_1[0].tolist()
Y_2 = df_2[0].tolist()
Y_3 = df_3[0].tolist()
X = range(1, len(Y_1)+1)

rcParams['lines.linewidth'] = 7

plt.figure(figsize=(14,9))

plt.plot(X, Y_1, linestyle=':', marker='^', markersize=14)
plt.plot(X, Y_2, linestyle='--', marker='^', markersize=14)
plt.plot(X, Y_3, linestyle='-', marker='o', markersize=18)
plt.legend(['p1', 'p2', 'p3'])
plt.xlabel('K', fontsize=45)
plt.title('Pair-counting Measures', fontsize=40)

plt.xticks(np.arange(1, 11, 1))
plt.yticks(np.arange(0.1, 1.1, 0.1))
plt.xlim([0.95, 10.05])
plt.ylim([-0.01, 1.01])

plt.savefig('plot/q2_5_4.png')
plt.show()
```
